In this project, we'll visualise the gender gap in college degrees based on data from the Department of Education Statistics in the US. This dataset contains the percentage of Bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category. 

We'll explore how we can communicate the nuanced narrative of gender gap using effective data visualisation. We'll also explore how to export the final diagram we create as an image file. This project will focus primarily on the aesthetics of the plots.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

# Create 3 different categories for degrees in the dataset, these lists are ordered in descending order by the percentage of degrees awarded to women.
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
# Set plot line colors to colorblind mode
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

fig = plt.figure(figsize =(18, 20))

# Generate first column of charts. STEM degrees
for sp in range(0, 6):
    # Going to have 17 plots total, so we'll use a 6 rows, 3 columns format for our plots.
    ax = fig.add_subplot(6, 3, 3 * sp + 1)
    
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c = cb_dark_blue, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[stem_cats[sp]], c = cb_orange, linewidth = 3, label = 'Men')
    
    # Remove the spines from the plots
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    # Remove the tick marks from the plot
    ax.tick_params(bottom = False, top = False, left = False, right = False)
    
    # Enable x-axis labels for the bottommost line chart in the STEM columns
    if sp == 5:
        ax.tick_params(labelbottom = True)
    else:
        ax.tick_params(labelbottom = False)
    
    # Set x, y limits
    ax.set_xlim(1970, 2011)
    ax.set_ylim(0, 100)
    
    # Set y tick values to show only 0 and 100
    ax.set_yticks([0, 100])
    
    # Add a horizontal line at y = 50 position, with a transperancy of 0.3 and a light gray colour in the Colour Blind 10 palette
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    
    # Set title of each plot
    ax.set_title(stem_cats[sp])
    
    # Add text to the plots at given positions
    if sp == 0:
        ax.text(2008, 13, 'Men')
        ax.text(2005, 82, 'Women')
    elif sp == 5:
        ax.text(2008, 85, 'Men')
        ax.text(2005, 10, 'Women')
    
    
# Generate second column of charts. Liberal Arts degrees
for sp in range(0, 5):
    ax = fig.add_subplot(6, 3, 3 * sp + 2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c = cb_dark_blue, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[sp]], c = cb_orange, linewidth = 3, label = 'Men')
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    ax.tick_params(bottom = False, top = False, left = False, right = False)
    
    if sp == 4:
        ax.tick_params(labelbottom = True)
    else:
        ax.tick_params(labelbottom = False)
    
    ax.set_xlim(1970, 2011)
    ax.set_ylim(0, 100)
    ax.set_yticks([0, 100])
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    ax.set_title(lib_arts_cats[sp])

    if sp == 0:
        ax.text(2008, 20, 'Men')
        ax.text(2005, 75, 'Women')
    
# Generate third column of charts. Other degrees
for sp in range(0, 6):
    ax = fig.add_subplot(6, 3, 3 * sp + 3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c = cb_dark_blue, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[other_cats[sp]], c = cb_orange, linewidth = 3, label = 'Men')
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.tick_params(bottom = False, top = False, left = False, right = False)
    
    if sp == 5:
        ax.tick_params(labelbottom = True)
    else:
        ax.tick_params(labelbottom = False)
    
    ax.set_xlim(1970, 2011)
    ax.set_ylim(0, 100)
    ax.set_yticks([0, 100])
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    ax.set_title(other_cats[sp])
    
    if sp == 0:
        ax.text(2008, 5, 'Men')
        ax.text(2005, 88, 'Women')
    elif sp == 5:
        ax.text(2008, 65, 'Men')
        ax.text(2005, 35, 'Women')

# Export the figure containing all the line charts.
plt.savefig('gender_degrees.png')

plt.show()
```

From looking through the plots, a number of findings can be determined.

**STEM Fields:**

For the Psychology field, the majority is women although it looks like there is still a considerable gender gap in the Engineering and CS fields. We can also see that Physical Sciences and Biology has become more popular with women over the years of this dataset. We can also see that the Maths and Statistic fields have been pretty similar in their low gender gap through the years. 

**Liberal Arts Fields:**

For most of the fields chosen in this dataset the majority of most were women. We can see no gender gap when it comes to Social Sciences and History.

**Other Fields:**

Fields such as Health Professions, Public Administration and Education have women as the majority. We can also see that fields such as Business, Architecture and Argriculture have become more popular with women over the years.
